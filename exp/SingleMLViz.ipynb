```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
```

```python
p_names = ['Ra', 'gpas', 'ffact']
df = pd.read_csv("/Users/admin/PROJECTS/SPE/parameter-inference/exp/exp_160526c/single_plots/ML.csv", names=p_names)

df.head()
```

```python
sb.pairplot(df, kind="kde")
```

```python
g = sb.pairplot(df, diag_kind='kde')
g.map_lower(sb.kdeplot, levels=5, color=".2")
g.map_upper(sb.kdeplot, levels=5, color=".2")
```

```python
data = df.to_numpy()
f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12,6))
ax1.plot(data[:, 0], label="Ra")
ax1.legend()
ax2.plot(data[:, 1], label="gpas")
ax2.legend()
ax3.plot(data[:, 2], label="ffact")
ax3.legend()
f.suptitle("For colored noise")
f.supxlabel("Repetition num")

plt.show()
```

```python
df_w = pd.read_csv("/Users/admin/PROJECTS/SPE/parameter-inference/exp/exp_160526w/single_plots/ML.csv", names=p_names)
df_cw = pd.read_csv("/Users/admin/PROJECTS/SPE/parameter-inference/exp/exp_160526cw/single_plots/ML.csv", names=p_names)
```

```python
data = df_w.to_numpy()
f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12,6))
ax1.plot(data[:, 0], label="Ra")
ax1.legend()
ax2.plot(data[:, 1], label="gpas")
ax2.legend()
ax3.plot(data[:, 2], label="ffact")
ax3.legend()
f.suptitle("For white noise")
f.supxlabel("Repetition num")

plt.show()
```

```python
data = df_cw.to_numpy()
f, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12,6))
ax1.plot(data[:, 0], label="Ra")
ax1.legend()
ax2.plot(data[:, 1], label="gpas")
ax2.legend()
ax3.plot(data[:, 2], label="ffact")
ax3.legend()
f.suptitle("For colored and white noise")
f.supxlabel("Repetition num")

plt.show()
```

```python
df["model"] = "c"
df_w["model"] = "w"
df_cw["model"] = "cw"

df_all = pd.concat([df, df_w, df_cw], ignore_index=True)
```

```python
g = sb.pairplot(df_all, hue="model", diag_kind='kde')
g.map_lower(sb.kdeplot, levels=5)
g.map_upper(sb.kdeplot, levels=5)
```

```python

df_cs = pd.concat([df, df_cw], ignore_index=True)
plt.figure(figsize=(16,16))
g = sb.pairplot(df_cs, hue="model", diag_kind='kde')
g.map_lower(sb.kdeplot, levels=5)
g.map_upper(sb.kdeplot, levels=5)
```
