```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data = np.genfromtxt('15312011-noise 1_comp_ON_soma_autocorr.txt')
data.shape
```

```python
plt.figure(figsize=(12,7))
plt.title('Data to be fitted')
plt.xlabel('Time (ms)')
plt.ylabel('Amplitude')
plt.plot(data[800001: 850001, 0],data[800001: 850001, 1])
plt.show()
```

```python
x = data[800001: 850001, 0]
y = data[800001: 850001, 1]
```

# Fit exponential to data

```python
from scipy.optimize import curve_fit
```

```python
def func(t, D, lamb):
    return D*lamb*np.exp(-t*lamb)

p_init = [30, 0.1]

p_opt, p_cov = curve_fit(func, x, y, p0=p_init)
p_err = np.sqrt(np.diag(p_cov))
print "Optimized parameters: " + str(p_opt)
print "Error of the fitting: " + str(p_err)
```

```python
tt = np.linspace(0,1000,50000)

plt.figure(figsize=(12,8))
plt.grid()
plt.title('Fit exponential to autocorrelation of noise')
plt.xlabel('Time [ms]')
plt.ylabel('Amplitude')
plt.plot(x,y, label = "autocorrelation", linewidth = 2, color = "#e45664", alpha = 1)
plt.plot(tt, func(tt, *p_opt), color='#56E4D6', linestyle="dashed", label = "fitted exponential\nD=%f lamb=%f" %(p_opt[0], p_opt[1]), linewidth = 2, alpha = 1)
plt.legend(loc="best")
plt.savefig("exponential_fit1.pdf")
plt.show()
```
